```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/kaggle/input/heart-disease-indicators/heart_disease_health_indicators.csv')

# Set the number of bins and range for the X-axis (age)
num_bins = 100  # You can adjust this number to control the number of points on the X-axis
age_range = (df['Age'].min(), df['Age'].max())

# Create a histogram of the 'BMI' column
plt.hist(df['BMI'], bins=num_bins, range=age_range)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('BMI Distribution by Age')
plt.show()
```
